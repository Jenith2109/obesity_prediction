```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from imblearn.combine import SMOTEENN
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, IsolationForest
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import plotly.graph_objects as go

#  X and y are already loaded
df = pd.DataFrame(X)
if isinstance(y, pd.DataFrame):
    y = y.iloc[:, 0]
else:
    y = pd.Series(y)

# Separate categorical columns and numerical columns
categorical_columns = ['Gender', 'CAEC', 'CALC', 'MTRANS', 'family_history_with_overweight', 'FAVC', 'SMOKE', 'SCC']
numerical_columns = [col for col in df.columns if col not in categorical_columns]

# Convert categorical columns to numeric
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# Encode target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df, y_encoded, test_size=0.20, random_state=1)

non_numeric_columns = X_train.select_dtypes(include=['object']).columns
if len(non_numeric_columns) > 0:
    raise ValueError(f"Non-numeric columns detected after encoding: {non_numeric_columns}")

# Summarize the shape of the training dataset
print(f"Training set shape: X_train: {X_train.shape}, y_train: {y_train.shape}")

# Show initial confusion matrix (before outlier removal)
initial_cm = confusion_matrix(y_test, y_test)  # Using y_test predictions for initial CM, adjust as necessary
plt.figure(figsize=(8, 6))
sns.heatmap(initial_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Initial Confusion Matrix (Before Outlier Removal)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

#  outliers in the training dataset
iso = IsolationForest(contamination=0.05, random_state=1)
yhat = iso.fit_predict(X_train)

# rows that are not outliers
mask = yhat != -1
X_train_iso, y_train_iso = X_train[mask], y_train[mask]

# Summarize the shape of the updated training dataset
print(f"Filtered training set shape: X_train_iso: {X_train_iso.shape}, y_train_iso: {y_train_iso.shape}")

# Scale the data (for the outlier-filtered dataset)
scaler = StandardScaler()
X_train_iso_scaled = scaler.fit_transform(X_train_iso)
X_test_scaled = scaler.transform(X_test)

# Apply SMOTE-ENN separately on the scaled data
smote_enn = SMOTEENN(random_state=1)
X_train_resampled, y_train_resampled = smote_enn.fit_resample(X_train_iso_scaled, y_train_iso)

# Summarize the shape of the balanced training dataset
print(f"SMOTE-ENN balanced training set shape: X_train_resampled: {X_train_resampled.shape}, y_train_resampled: {y_train_resampled.shape}")

# Define models
models = {
    'Random Forest': RandomForestClassifier(random_state=1),
    'Gradient Boosting': GradientBoostingClassifier(random_state=1),
    'Support Vector Machine': SVC(random_state=1),
    'Logistic Regression': LogisticRegression(random_state=1, max_iter=1000)
}

# Define hyperparameter grids for each model
param_grids = {
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'max_features': ['sqrt', 'log2', None]
    },
    'Gradient Boosting': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7],
        'subsample': [0.8, 0.9, 1.0]
    },
    'Support Vector Machine': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    },
    'Logistic Regression': {
        'C': [0.1, 1, 10],
        'solver': ['liblinear', 'lbfgs']
    }
}

# Train and evaluate each model
results = []
confusion_matrices = {}

for model_name, model in models.items():
    print(f"Training and evaluating {model_name}...")

    # Grid Search for best hyperparameters
    grid_search = GridSearchCV(model, param_grids[model_name], cv=StratifiedKFold(5), scoring='accuracy')
    grid_search.fit(X_train_resampled, y_train_resampled)

    # Best model from grid search
    best_model = grid_search.best_estimator_

    # Cross-validate the best model
    cv_scores = cross_val_score(best_model, X_train_resampled, y_train_resampled, cv=StratifiedKFold(5), scoring='accuracy')
    train_accuracy = cv_scores.mean()

    # Predict on the test set
    y_test_pred = best_model.predict(X_test_scaled)

    # Evaluate predictions
    test_accuracy = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred, average='weighted')
    recall = recall_score(y_test, y_test_pred, average='weighted')
    f1 = f1_score(y_test, y_test_pred, average='weighted')

    results.append({
        'Model': model_name,
        'Training Accuracy (CV)': train_accuracy,
        'Test Accuracy': test_accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    })

    # Store confusion matrix
    confusion_matrices[model_name] = confusion_matrix(y_test, y_test_pred)

#  visualization
df_results = pd.DataFrame(results)

# Print results in a table
fig = go.Figure(data=[go.Table(
    header=dict(values=list(df_results.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[df_results[col] for col in df_results.columns],
               fill_color='lavender',
               align='left'))
])
fig.update_layout(title='Model Performance Metrics')
fig.show()

# Visualization of results
df_results.set_index('Model').plot(kind='bar', figsize=(14, 8), colormap='viridis')
plt.title('Comparison of Model Performance Metrics')
plt.ylim(0, 1)
plt.ylabel('Score')
plt.xlabel('Model')
plt.xticks(rotation=45)
plt.legend(loc='best')
plt.show()

# Preparing data for bar graph
df_results_long = df_results.melt(id_vars='Model', var_name='Metric', value_name='Score')

# Plotting the additional bar graph using Plotly Express
fig = px.bar(df_results_long, x='Model', y='Score', color='Metric', barmode='group',
             title='Comparison of Model Performance Metrics',
             labels={'Score': 'Score', 'Model': 'Model'},
             height=600)
fig.update_layout(legend=dict(title='Metric'))
fig.show()

# Confusion matrix for each model
for model_name, cm in confusion_matrices.items():
    fig = px.imshow(cm, text_auto=True, color_continuous_scale='Blues',
                    labels=dict(x='Predicted', y='Actual', color='Count'),
                    title=f'Confusion Matrix for {model_name}')
    fig.show()

# Overall confusion matrix
overall_cm = sum(confusion_matrices.values())
fig = px.imshow(overall_cm, text_auto=True, color_continuous_scale='Blues',
                labels=dict(x='Predicted', y='Actual', color='Count'),
                title='Overall Confusion Matrix')
fig.show()
```
